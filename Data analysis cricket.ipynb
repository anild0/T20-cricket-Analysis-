```python
import pandas as pd
import json
```

###



match result details



```python
with open (r"C:\Users\Denny\OneDrive\Desktop\data analysis cric\json files\t20_wc_match_results.json") as f:
    df=json.load(f)
df=pd.DataFrame(df[0]["matchSummary"])
```

```python
df=df.rename({"scorecard":"match_id"},axis=1)
```

```python
df2=df[["match_id","matchDate","team1","team2","winner","margin","ground"]]
```

```python
match_results=df2
```

```python
match_ids_dict = {}

for index, row in df2.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

###
Batting



```python
with open(r"C:\Users\Denny\OneDrive\Desktop\data analysis cric\json files\t20_wc_batting_summary.json") as f:
    df=json.load(f)
```

```python
all_runs=[]
for x in df:
    all_runs.extend(x['battingSummary'])
```

```python
df=pd.DataFrame(all_runs)
```

```python
df
```

```python
df["dismissal"]=df.dismissal.apply(lambda x:"out" if len(x)>0 else "not out")
```

```python
df
```

```python
def split(x):
    d=x.split("(")
    return d[0]
    
```

```python
df.batsmanName=df.batsmanName.apply(split)
```

```python
df
```

```python
df["match_id"]=df["match"].map(match_ids_dict)
```

```python
df1=df[["match_id","match","teamInnings","battingPos","batsmanName","dismissal","runs","balls","4s","6s","SR"]]
```

```python
df1
```

```python
df1.rename({"4s":"fours","6s":"six"},axis=1)
```

```python
batting=df1
```
